```python
#default dict

from collections import defaultdict
data = [{'category': 'A', 'value': 10}, {'category': 'B', 'value': 20}, {'category': 'A', 'value': 5}]

# 일반 dict
def normal_dict(data):
    summary = {}

    for item in data:
        if item['category'] not in summary:
            summary[item['category']] = 0
        summary[item['category']] += item['value']
    return summary

# defaultdict
def default_dict(data):
    summary_dd = defaultdict(int) # 값을 int(0)으로 초기화
    for item in data:
        summary_dd[item['category']] += item['value']

    return dict(summary_dd)
print(normal_dict(data))
print(default_dict(data))

```

```python
def apply_automation_rules_v2(campaigns, rules):
    # 입력받은 campaigns 리스트를 직접 수정합니다.
    for campaign in campaigns:
        for rule in rules:
            # 1. 규칙에서 필요한 값들을 변수로 추출
            field_to_check = rule['condition_field']
            target_value = rule['condition_value']
            action = rule['action']
            
            # 2. campaign이 해당 필드 값을 가지고 있는지 확인
            if field_to_check in campaign:
                campaign_value = campaign[field_to_check]
                
                # 3. 연산자(operator)에 따라 동적으로 조건 검사
                operator = rule.get('condition_operator', '==') # 연산자가 없으면 '=='를 기본값으로
                
                match = False
                if operator == '==':
                    if campaign_value == target_value:
                        match = True
                elif operator == '>':
                    if campaign_value > target_value:
                        match = True
                elif operator == '<':
                    if campaign_value < target_value:
                        match = True
                # (필요하다면 여기에 >=, <=, != 등 다른 연산자 추가)

                # 4. 조건이 일치하면 행동을 적용하고, 이 캠페인에 대한 규칙 검사는 중단
                if match:
                    campaign['status'] = action
                    break # 다음 캠페인으로 넘어감
                    
    return campaigns

# --- 테스트 실행 ---
# campaigns, rules 변수는 문제에 주어진 그대로 사용
updated_campaigns = apply_automation_rules_v2(campaigns, rules)
print(updated_campaigns)
```

```python
"""
문제 설명
광고 성과 기록(performance_logs)과 각 광고 소재의 메타데이터(creative_metadata)가 별도로 주어집니다.
performance_logs에 기록된 성과를 creative_metadata에 정의된 **라벨(label)**을 기준으로 집계하는 함수 aggregate_by_label을 작성하세요.

하나의 소재(creative_id)는 여러 개의 라벨을 가질 수 있으며, 이 경우 해당 소재의 성과는 관련된 모든 라벨에 중복으로 집계되어야 합니다.
"""

#입력 예시

# 1. 광고 성과 기록 (시간 순서 없이 들어옴)
performance_logs = [
    {'creative_id': 'c_001', 'clicks': 100, 'cost': 15000},
    {'creative_id': 'c_002', 'clicks': 120, 'cost': 18000},
    {'creative_id': 'c_003', 'clicks': 50, 'cost': 7500},
    {'creative_id': 'c_002', 'clicks': 130, 'cost': 20000}, # c_002의 추가 성과
    {'creative_id': 'c_001', 'clicks': 110, 'cost': 16000}, # c_001의 추가 성과
]

# 2. 광고 소재 메타데이터 (ID를 key로 가짐)
creative_metadata = {
    'c_001': {'name': 'blue_shirt.jpg', 'labels': ['fashion', 'model_shot', 'blue']},
    'c_002': {'name': 'red_pants.jpg', 'labels': ['fashion', 'product_shot', 'red']},
    'c_003': {'name': 'white_shoes.png', 'labels': ['fashion', 'product_shot', 'white', 'model_shot']},
}


#출력예시
#{
#    'fashion': {'total_clicks': 510, 'total_cost': 76500},
#    'model_shot': {'total_clicks': 260, 'total_cost': 38500},
#    'blue': {'total_clicks': 210, 'total_cost': 31000},
#    'product_shot': {'total_clicks': 300, 'total_cost': 45500},
#    'red': {'total_clicks': 250, 'total_cost': 38000},
#    'white': {'total_clicks': 50, 'total_cost': 7500}
#}
```

```python
#질문 : 문제의 상황은 명확합니다. 광고의 성과 기록과 메타데이터가 들어오면 메타데이터의 label을 바탕으로 성과의 cost와 click을 집계하는 것입니다. 이때 성과는 중복되어 집계되어야 합니다.
#질문 : performance_log에 존재하는 creative_id가 metadata에 존재한다는 보장이 필요합니다. 존재하지 않을 수도 있다면 그때의 KeyError를 처리하는 부분을 추가하겠습니다

#문제해결 전략
#defaultdict를 사용하여 total_clicks와 total_cost를 선언한 후 라벨이 존재한다면 기존 라벨에 cost와 clicks를 더하고 존재하지 않는다면 생성 후 더하는 방식으로 진행하겠습니다.

from collections import defaultdict

def aggregate_by_label(performance_logs, creative_metadata):
    label_agg = defaultdict(lambda: {'total_clicks':0, 'total_cost':0})
    for log in performance_logs:
        metadata = creative_metadata[log['creative_id']]
        label_list = metadata['labels']
        for label in label_list:
            label_agg[label]['total_clicks'] += log['clicks']
            label_agg[label]['total_cost'] += log['cost']

    return dict(label_agg)

print("------------------------------test case------------------------------")        
print(aggregate_by_label(performance_logs, creative_metadata))

#개선점 및 추가 논의사항: 한가지 광고를 통해 중복되는 성과가 기록됩니다. 즉 실재 광고의 총 성과에 비해 그 결과가 부풀려지는 결과가 발생할 수 있어 보입니다. 광고의 성과가 어느쪽으로 편향되어있는지를 판단할 수 있는 로직을 추가한다면 더욱 실재적인 서비스로 작용할 수 있어보입니다.
```
